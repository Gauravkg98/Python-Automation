```python
import statistics

Folder ="F://Swimdat//"



def read_swim_data(filename):
  
  swimmer, age, distance, stroke = filename.removesuffix(".txt").split("-")

  with open (Folder+filename) as file:
    lines = file.readlines()
    times = lines[0].strip().split(",")
  convert = []
  for t in times:
    if ":" in t:
      minutes, rest = t.split(":")
      seconds, hunderth = rest.split(".")
      convert.append((int(minutes)*60*100)+(int(seconds)*100)+int(hunderth))
    else:
      mintues = 0
      seconds,hunderth = t.split(".")
    convert.append((int(minutes)*60*100)+(int(seconds)*100)+int(hunderth))

  average=statistics.mean(convert)
  min_secs, hunderths =str(round(average/100,2)).split(".") 
  min_secs = int(min_secs)
  minutes,sec = str((min_secs /60)).split(".")
  average = str(minutes)+" : "+str(sec)+" . "+str(hunderths)
  
  

  return f"""Swimmer Name = {swimmer}'\n' Age ={ age}'\n' Stroke = {stroke}'\n' Distance= {distance}'\n' Times= {times}'\n'Average= {average}"""

filename = "Darius-13-100m-FLY.txt"
print( read_swim_data(filename))
```
